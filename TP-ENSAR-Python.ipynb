```python
# Lecture du fichier de donnnee
df =pd.read_csv("salaries.csv",sep=";")
df.head(n=10)
```

```python
print(df.columns)
```

```python
# Spécifier jsute le jeu de données 
ggplot(data =df)
```

```python
# Spécifier le jeu de données et la cartographie sans la géométrie
ggplot(df,aes(x = "yrs.service", y = "salary"))
```

```python
#Tracer un nuage de points avec Plotnine
ggplot(df,aes(x = "yrs.service", y = "salary"))+geom_point()
```

```python
#des graphiques type histogramme
ggplot(df,aes(x = "yrs.service",y="salary"))+geom_line()
```

```python
#Bar chart
ggplot(data =df,mapping = aes(x= "yrs.service"))+geom_bar()
```

```python
#Traçage d'un diagramme en boîte avec Plotnine
ggplot(df,aes(x = "yrs.service",y="salary"))+geom_boxplot()
```

```python
#Bar chart
ggplot(data =df,mapping = aes(x = "yrs.service"))+geom_bar(fill = "red",color="black")
```

```python
#Labelisation
ggplot(data =df,mapping = aes(x = "yrs.service"))+geom_bar(fill = "cornflowerblue",color="black")+ labs(x = "Service",y = "Frequence",title = "Participants")
```

```python
#Tracer un nuage de points et distinguer en couleur la geometrie
ggplot(df,aes(x = "yrs.service",y="salary"))+geom_point(size=0.2,color="red")
```

```python
##Rendre les points rouges, plus grands et semi-transparents
ggplot(data =df,mapping = aes(x = "yrs.service",y="salary"))+geom_point(size=1,color="orange",alpha=1)
```

```python
df["sex"].value_counts()
```

```python
#On peut distinguer dans la fonction esthétique aes
ggplot(data =df,mapping = aes(x = "yrs.service",y="salary",color="sex"))+geom_point(size=1,alpha=1)
```

```python
#Ajout de légende
ggplot(data =df,mapping = aes(x = "yrs.service",y="salary",color="sex"))+geom_point(size=1,alpha=1)+theme(legend_position=(0.9,0.91),legend_direction='vertical')
```

```python
from mizani.formatters import currency_format
```

```python
#Changement d"échelle des x et y et personalisation de couleurs
ggplot(data =df,mapping = aes(x = "yrs.service",y="salary",color="sex"))+geom_point(size=1,alpha=1)+scale_x_continuous(breaks = range(0,60,10))+scale_y_continuous(breaks=range(0,300000,100000),labels=currency_format())+scale_color_manual(values = ["indianred","cornflowerblue"])+ theme(legend_position=(0.85,0.8),legend_direction='vertical')
```
